```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
dataFrame = pd.read_csv('ArtPainting.csv')
pd.options.display.float_format = '{:.0f}'.format
```

```python
dataFrame = dataFrame[['name', 'country', 'birth', 'death', 'title', 'medium', 'size', 'signed','stamped','auction_house','lower_estimate','upper_estimate','actual_price','currency']]
```

```python
dataFrame.head(10)
```

```python
dataFrame['NumberOfYearsArtistsDied'] = dataFrame['death'] - dataFrame['birth']
```

```python
dataFrame['NumberOfYearsArtistsDied'] = 2017 - dataFrame['death']
dataFrame['NumberOfYearsArtistsBorned'] = 2017 - dataFrame['birth']
dataFrame['mean_estimate'] = (dataFrame['lower_estimate'] + dataFrame['upper_estimate'])/2
dataFrame.head(20)
```

```python
TotalPriceByArtists = dataFrame.groupby('name')['lower_estimate', 'upper_estimate','mean_estimate', 'actual_price'].sum()
```

```python
TotalPriceByArtists.sort_values('actual_price', ascending=False).head(20)
```

```python
dataFrame['ArtworkCount'] = dataFrame.groupby('name')['name'].transform('count')
```

```python
CountPaintingByArtists = dataFrame.groupby('name').size().reset_index(name="Total")
CountPaintingByArtists = CountPaintingByArtists.sort_values('Total', ascending=False)
```
